```python
def fib(n):
    a = 0
    b = 1

    for i in range(n):
        print(a, end=" ")
        c = a+b
        a = b
        b = c

n = int(input("Enter the number of terms "))
fib(n)
```

```python
def myPow(x, n):
    result = 1
    count = 0

    while count < n:
        result = result * x
        count += 1

    return result


x = int(input("Enter base (x): "))
n = int(input("Enter power (n): "))

print("Answer:", myPow(x, n))
```

```python
def SumDigits(n):
    total = 0

    while n > 0:
        digit = n % 10      # get last digit
        total += digit      # add it to total
        n = n // 10         # remove last digit

    print("Sum of digits:", total)


# Take input from user
num = int(input("Enter a number: "))
SumDigits(num)
```

```python
def squaredSum(arr):
    total = 0
    i = 0

    while i < len(arr):
        total += arr[i] * arr[i]   # square and add
        i += 1

    return total


# Example
nums = [2, 3, 4]
print("Sum of squares:", squaredSum(nums))
```

```python
def sumDigits(num):
    total = 0
    while num > 0:
        total += num % 10
        num //= 10
    return total

# take input from user
num = int(input("Enter a number: "))

# call function and print result
result = sumDigits(num)
print("Sum of digits:", result)
```
